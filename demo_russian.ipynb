### Algorithm description part   

1. split text to sentences
2. for each sentence apply wordrank algorithm
    - vectorize each word in sentence
    - make adjacency matrix
    - make weighted graph from that matrix
    - find all cliques of graph
    - find clique with max sum of weights on edges --> list of words
3. build similarity matrix
4. make weighted graph from similarity matrix
5. find all cliques of that graph
6. choose the first clique from the list
7. find match between indexes of nodes of the clique and original sentences --> print original sentences
